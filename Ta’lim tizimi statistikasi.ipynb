# **Oâ€˜zbekiston Taâ€™lim Tizimi Statistikasi (Maktablar, OTMlar, Talabalar soni)**
## ðŸ“š Kirish
Ushbu tahlil Oâ€˜zbekiston Respublikasi taâ€™lim sohasining asosiy statistik koâ€˜rsatkichlarini (maktablar, oliy ta'lim muassasalari, o'quvchi va talabalar soni) oâ€˜z ichiga oladi. Maâ€™lumotlar simulyatsiya qilingan.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ma'lumotlarni yaratish
data = {
    'Viloyat': ['Andijon', 'Buxoro', 'Jizzax', 'Qashqadaryo', 'Navoiy', 'Namangan', 'Samarqand', 'Surxondaryo', 'Sirdaryo', 'Toshkent v.', 'Farg\'ona', 'Xorazm', 'Toshkent sh.', 'Qoraqalpog\'iston'],
    'Maktablar_soni': [780, 560, 450, 910, 390, 810, 950, 720, 310, 750, 974, 520, 300, 700],
    'Oquvchilar_soni': [500000, 380000, 290000, 650000, 250000, 520000, 780000, 450000, 190000, 480000, 680000, 350000, 550000, 455752],
    'Pedagoglar_soni': [30000, 22000, 18000, 40000, 15000, 31000, 43333, 27000, 11000, 29000, 40235, 20000, 25000, 24000],
    'OTM_soni': [10, 7, 6, 13, 5, 9, 15, 8, 4, 7, 11, 7, 61, 5],
    'OTM_talabalari': [71926, 45000, 38000, 81711, 28000, 62000, 107693, 55000, 25000, 49000, 94814, 40000, 451457, 35314]
}
df = pd.DataFrame(data)
df['Oquvchi_pedagog_nisbati'] = df['Oquvchilar_soni'] / df['Pedagoglar_soni']

# OTM talabalarining o'sish dinamikasi
talaba_osish_df = pd.DataFrame({
    'Yil': [2021, 2022, 2023, 2024],
    'Jami_talabalar': [550383, 715000, 950000, 1184915]
})
print("Ma'lumotlar muvaffaqiyatli yuklandi.")
```

## **O'quvchi/Pedagog Nisbati (Yuklama) Grafiki**

```python
# Eng yuqori 5 viloyat bo'yicha o'quvchi/pedagog nisbati
nisbat_df = df.sort_values(by='Oquvchi_pedagog_nisbati', ascending=False).head(5)

plt.figure(figsize=(10,6))
plt.bar(nisbat_df['Viloyat'], nisbat_df['Oquvchi_pedagog_nisbati'], color='skyblue')
plt.title("Eng Yuqori O'quvchi/Pedagog Nisbatiga ega Viloyatlar")
plt.xlabel("Viloyat")
plt.ylabel("O'quvchi/Pedagog Nisbati")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

## **OTM Talabalar Soniga Ko'ra Eng Yirik 5 Ta Viloyat**

```python
otm_df = df.sort_values(by='OTM_talabalari', ascending=False).head(5)
plt.figure(figsize=(10,6))
plt.bar(otm_df['Viloyat'], otm_df['OTM_talabalari']/1000, color='indianred')
plt.title("Eng Ko'p Talabalar Soniga Ega Viloyatlar (Ming nafar)")
plt.xlabel("Viloyat")
plt.ylabel("Talabalar soni (Ming nafar)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

## **OTM Talabalarining Yillik O'sish Dinamikasi**

```python
plt.figure(figsize=(9,5))
plt.plot(talaba_osish_df['Yil'], talaba_osish_df['Jami_talabalar']/1000, marker='o', linestyle='-', color='darkgreen')
plt.title("OTM Talabalarining Yillar Bo'yicha O'sishi")
plt.xlabel("Yil")
plt.ylabel("Jami talabalar soni (Ming nafar)")
plt.xticks(talaba_osish_df['Yil'])
plt.grid(axis='y', linestyle=':')
plt.tight_layout()
plt.show()
```
